```python
#Find Max in List (Without max())
numbers = [10, 45, 3, 78, 22]

max_num = numbers[0]

for num in numbers:
    if num > max_num:
        max_num = num

print("Maximum:", max_num)
```

```python
#count Frequency
numbers = [1,2,2,3,3,3,4]

count = 0

for num in numbers:
    if num == 3:
        count += 1

print("Frequency:", count)
```

```python
customers = [
    {"name": "A", "spending": 5000},
    {"name": "B", "spending": 8000},
    {"name": "C", "spending": 3000}
]
#Process Data Using Loop
total_revenue = 0

for customer in customers:
    total_revenue += customer["spending"]

print("Total Revenue:", total_revenue)

#Classify Customers
for customer in customers:
    if customer["spending"] > 6000:
        print(customer["name"], "is Premium")
    else:
        print(customer["name"], "is Regular")
```

```python
#COMPARE ALL NUMBERS
numbers = [1, 2, 3]

for i in numbers:
    for j in numbers:
        print(i, j)
```

```python
numbers = [2, 7, 11, 15]
target = 9

seen = {}

for num in numbers:
    diff = target - num
    if diff in seen:
        print("Pair found:", diff, num)
    seen[num] = True
```
